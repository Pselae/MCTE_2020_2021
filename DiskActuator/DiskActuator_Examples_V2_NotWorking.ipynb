```python
import numpy as np
import matplotlib.pyplot as mpl
%config InlineBackend.figure_formats = ['svg']
```

```python
C_T_lst = []
C_P_lst = []
η_lst = []

B = 0.05
Fr1 = 0.05

for Fr4b in np.linspace( Fr1*1.001, Fr1*1.6, 1000 ):

  ζ4 = (1/2.)*Fr1**2 - 1/2.*Fr4b**2 + 1
  
  Fr4t = (Fr1 - Fr4b*ζ4 + np.sqrt(B**2*Fr4b**2 - 2*B*Fr1**2 + 2*B*Fr1*Fr4b \
          + B*ζ4**2 - B + Fr1**2 - 2*Fr1*Fr4b*ζ4 + Fr4b**2*ζ4**2))/B
  ζ4b = (Fr1 - Fr4t*ζ4)/(Fr4b - Fr4t)
  ζ4t = (-Fr1 + Fr4b*ζ4)/(Fr4b - Fr4t)
  
  Fr2t = Fr4t*ζ4t/B
  C_T = (Fr4b**2 - Fr4t**2)/Fr1**2
  C_P = C_T*Fr2t/Fr1

  if C_P <= 0.0: break

  C_4 = Fr4b**2*ζ4b + Fr4t**2*ζ4t + (1/2)*ζ4**2
  C_5 = (27*Fr1**2 + (1/2)*np.sqrt(-864*C_4**3 + 2916*Fr1**4))**(1/3)
  ζ5 = -2*C_4/C_5 - 1/3*C_5
  Fr5 = Fr1 / ζ5

  C_Ptotal = 2*((1/2)*Fr1**2 - 1/2*Fr5**2 - ζ5 + 1)/(B*Fr1**2)
  η = C_P/C_Ptotal

  η_lst.append( η )
  C_P_lst.append( C_P )
  C_T_lst.append( C_T )

mpl.plot( C_T_lst, C_P_lst, label="$\mathrm{Fr}_1=%.2f$" % Fr1 )
mpl.plot( C_T_lst, η_lst, label="$\mathrm{Fr}_1=%.2f$" % Fr1 )
mpl.xlabel( "$C_T$" )
mpl.ylabel( "$C_P$" )
mpl.grid()
mpl.legend();
```

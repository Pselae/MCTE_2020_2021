```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as mpl
%config InlineBackend.figure_formats = ['svg']
mpl.rcParams["figure.figsize"] = (12, 3)
```

```python
df = pd.read_excel( 'Data.xlsx', sheet_name='Sheet1', header=0, skiprows=(1,) )
df
```

```python
def plot_time_series( N ):
  Œ∂1 = df['ùúÅ1'][N]
  Œ∂2 = df['ùúÅ2'][N]
  œâ1 = df['ùúî1'][N]
  œâ2 = df['ùúî2'][N]
  œÜ1 = df['ùúô1'][N]
  œÜ2 = df['ùúô2'][N]

  name = df['Name'][N]
  tide_period = 1.0 / ( œâ2 - œâ1 ) * 2.0*np.pi

  t = np.linspace( 0, tide_period, 1000 )
  A = Œ∂1 * np.cos( œâ1 * t + œÜ1 ) + Œ∂2 * np.cos( œâ2 * t + œÜ2 )

  mpl.plot( t, A )
  mpl.title( name );
```

```python
plot_time_series( 0 )
```

```python
plot_time_series( 1 )
```
